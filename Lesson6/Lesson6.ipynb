## Practical 1

Create a script that calculates the kinematics of a cannonball:
* Prompt the user for height, velocity and angle
* Print time-of-flight
* Plot the x,y trajectory
* Print impact location (x,y)

HINT:

$x(t) = x_0 + v_0 \cdot t + 1/2 \cdot a \cdot t^2$

```python
# 1.
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0.,10.,0.02)
h = float(input("What is the initial height of the object: "))
v = float(input("What is the initial velocity of the object: "))
theta = float(input("What is its angle from zero: "))


theta = theta *(np.pi)/180

def x(time,angle,speed):
    x = time * speed * np.cos(angle)
    return(x)

def y(time, angle, speed, height):
    y = height + speed * np.sin(angle) * time - 4.9 * time**2
    for i in y:
        if i <= 0:
            i = 0
    return(y)
   

plt.figure(1, figsize = (12,12))
plt.subplot(211)
plt.plot(t,y(t,theta,v,h),'k')
plt.title("height with time")
    
plt.subplot(212)
plt.plot(t,x(t,theta,v),'k')
plt.title("distance with time")

plt.show()
    

```

## Practical 2

Create a script to play rock-paper-scissors:
* Prompt the user a choice
* Print the result based off computer choice

Hint: Dictionaries can help turn strings into numbers for comparison.

```python
# 2.

from random import sample

answer = "yes"

while answer == "yes":
    user = ("none")
    while user == "none":
        user = (input("Input 'rock','paper', or 'scissors': "))
        if user != "rock" and user != "paper" and user != "scissors":
            user = "none"
        
    list = ["rock","paper","scissors"]

    computer = sample(list,1)
    comp2 = computer[0]


    winvalue = "null"

    if user == comp2:
        winvalue = "Tied!!!"
    if user == "rock" and comp2 == "scissors" or user == "scissors" and comp2 == "paper" or user == "paper" and comp2 == "rock":
        winvalue = "Won!!!"
    if user == "rock" and comp2 == "paper" or user == "scissors" and comp2 == "rock" or user == "paper" and comp2 == "scissors":
        winvalue = "Lost!!!"
    print()
    print("You have chosen", user, "and the computer has chosen", comp2)
    print()
    print("It looks like you have", winvalue)
    print()
    answer = input("Would you like to play again? (yes or no): ")
    print()
    
print("Okay, goodbye, friend :( ")
```
