Introduction to PANDAS

Pandas is an open source, library. It is developed for high-performance data analysis and easy-to-use data structures.

Pandas itself is built for use with the Python programming language.

```python
#This is the usual load order, and if you are doing data analysis of any kind these are the common libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

The primary structure for manipulating and processing any data in the Pandas environment is the **DataFrame**

This is where your data will sit, and the basis for all manipulationa and analysis. 

DataFrames can be created through several methods:

-Read in a previous data storage file (csv, pickle, etc..)

-Assign through a dictionary of 1D ndarrays, lists, ext...

-Structured or record ndarray

-Calling from another *DataFrame*

```python
'''
From a previous storage file
'''

#call using read_(appropriate file structure)
df0 = pd.read_csv('wages_hours.csv', sep = '\t')
df0
```

```python
"""
From assignment of a dictionary
"""
#create dictionary
d = {'one' : pd.Series([1.,2.,3.], index=['a','b','c']), 'two' : pd.Series([1.,2.,3.,4.],index=['a','b','c','d'])}
#assign that dictionary to a DataFrame
df1 = pd.DataFrame(d)
#call/output the dataframe
df1
```

```python
df1.index, df1.columns
```

```python
"""
From structure or record array
"""
#Create array of zeros, type assigned
data =np.zeros((2,), dtype=[('A','int'),('B','float'),('C','a5')])
#fill accordingly
data[:] = [(1,2.,'Hello'), (2,3.,"World")]
#assign to DataFrame
df2 = pd.DataFrame(data, index=['first','second'])
#call/output
df2
```

```python
"""
From simply copying another array, or a slice of another array
"""
df3 = df0.loc[:4,'HRS':'AGE'] #Slice first 5 rows, including columns 'HRS' to 'AGE'
df3
```

One thing I always find helpful is the Boolean Selection built-in to the DataFrame description:

I can use Boolean discriptions to only look at data I'm interested in

```python
'''
Select only those indices for relevant values in the DataFrame 
whose 'ASSET' column entry is greater than or equal to the mean

Then display only those rows using .iloc
'''

Selection = df3[df3['ASSET'] >= df3['ASSET'].mean()].index
df3.iloc[Selection]
```

I can also use simple mathematical operations to assign new columns to my DataFrame

```python
df3['New Column'] = df3['NEIN'] * df3['RATE'].mean()
df3
```

I can apply functions, either defined above or with lambda functions

```python
df3['NEIN'].apply(lambda x: x**2 + 4*x)
```
